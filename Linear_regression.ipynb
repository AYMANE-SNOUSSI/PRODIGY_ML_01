```python
data['LotFrontage'].fillna(data['LotFrontage'].median(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = data[['LotArea','FullBath','BedroomAbvGr']]
y = data['SalePrice']
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
tse = mean_squared_error(Y_test, predictions)
print(f"mean squared error : {tse}")
```

```python
test_data = pd.read_csv("test.csv")
```

```python
X_test_actual = test_data[['LotArea','FullBath','BedroomAbvGr']]
```

```python
test_prediction = model.predict(X_test_actual)
```

```python
print(test_prediction)
```

```python
output = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': test_prediction})
```

```python
output.to_csv("submission.csv", index=False)
```
